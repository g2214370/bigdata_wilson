```python
fields = [
    u"觀測時間",
    u"測站氣壓",
    u"海平面氣壓",
    u"測站最高氣壓",
    u"測站最高氣壓時間",
    u"測站最低氣壓",
    u"測站最低氣壓時間",
    u"氣溫",
    u"最高氣溫",
    u"最高氣溫時間",
    u"最低氣溫",
    u"最低氣溫時間",
    u"露點溫度",
    u"相對溼度",
    u"最小相對溼度",
    u"最小相對溼度時間",
    u"風速",
    u"風向",
    u"最大陣風",
    u"最大陣風風向",
    u"最大陣風風速時間",
    u"降水量",
    u"降水時數",
    u"10分鐘最大降水量",
    u"10分鐘最大降水起始時間",
    u"一小時最大降水量",
    u"一小時最大降水量起始時間",
    u"日照時數",
    u"日照率",
    u"全天空日射量",
    u"能見度",
    u"A型蒸發量",
]
```

```python
def fetchDate(url):
    rsps = urllib.urlopen( url.encode(u"utf8") )
    soup = BeautifulSoup(rsps.read())
    
    pd_month = pd.DataFrame(columns=fields)
    
    cnt =0
    for tr in soup.find_all("tr"):
        cnt += 1
        if cnt >3:
            
            rows = [ y.get_text().strip() if not y.get_text().strip() in [u"T", u"X"]else "" for y in tr.find_all("td") ]
        
            rows[0] = int(rows[0]) if len(rows[0]) > 0 else 0.
            rows[1] = float(rows[1]) if len(rows[1]) > 0 else 0.
            rows[2] = float(rows[2]) if len(rows[2]) > 0 else 0.
            rows[3] = float(rows[3]) if len(rows[3]) > 0 else 0.
            rows[4] = datetime.strptime(rows[4].split(" ")[0],'%Y-%m-%d') if len(rows[4]) > 0 else 0.
            rows[5] = float(rows[5]) if len(rows[5]) > 0 else 0.
            rows[6] = datetime.strptime(rows[6].split(" ")[0],'%Y-%m-%d') if len(rows[6]) > 0 else 0.
            rows[7] = float(rows[7]) if len(rows[7]) > 0 else 0.
            rows[8] = float(rows[8]) if len(rows[8]) > 0 else 0.
            rows[9] = datetime.strptime(rows[9].split(" ")[0],'%Y-%m-%d') if len(rows[9]) > 0 else 0.
            rows[10] = float(rows[10]) if len(rows[10]) > 0 else 0.
            rows[11] = datetime.strptime(rows[11].split(" ")[0],'%Y-%m-%d') if len(rows[11]) > 0 else 0.
            rows[12] = float(rows[12]) if len(rows[12]) > 0 else 0.
            rows[13] = float(rows[13]) if len(rows[13]) > 0 else 0.
            rows[14] = float(rows[14]) if len(rows[14]) > 0 else 0.
            rows[15] = datetime.strptime(rows[15].split(" ")[0],'%Y-%m-%d') if len(rows[15]) > 0 else 0.
            rows[16] = float(rows[16]) if len(rows[16]) > 0 else 0.
            rows[17] = float(rows[17]) if len(rows[17]) > 0 else 0.
            rows[18] = float(rows[18]) if len(rows[18]) > 0 else 0.
            rows[19] = float(rows[19]) if len(rows[19]) > 0 else 0.
            rows[20] = datetime.strptime(rows[20].split(" ")[0],'%Y-%m-%d') if len(rows[20]) > 0 else 0.
            rows[21] = float(rows[21]) if len(rows[21]) > 0 else 0.
            rows[22] = float(rows[22]) if len(rows[22]) > 0 else 0.
            rows[23] = float(rows[23]) if len(rows[23]) > 0 else 0.
            rows[24] = datetime.strptime(rows[24].split(" ")[0],'%Y-%m-%d') if len(rows[24]) > 0 else 0.
            rows[25] = float(rows[25]) if len(rows[25]) > 0 else 0.
            rows[26] = datetime.strptime(rows[26].split(" ")[0],'%Y-%m-%d') if len(rows[26]) > 0 else 0.
            rows[27] = float(rows[27]) if len(rows[27]) > 0 else 0.
            rows[28] = float(rows[28]) if len(rows[28]) > 0 else 0.
            rows[29] = float(rows[29]) if len(rows[29]) > 0 else 0.
            rows[30] = float(rows[30]) if len(rows[30]) > 0 else 0.
            rows[31] = float(rows[31]) if len(rows[31]) > 0 else 0.
            pd_month.loc[cnt-3] = rows
    return pd_month
```
